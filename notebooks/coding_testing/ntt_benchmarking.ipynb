```python
import torch

from mlkem_gpu.device import DEVICE
from mlkem_gpu.params import KYBER_N

from mlkem_gpu.ntt_gpu import ntt_coeffs
from mlkem_gpu.poly_gpu import Poly, poly_ntt, poly_ntt, poly_invntt_tomont
```

```python
p = Poly(torch.arange(256, dtype=torch.int16, device=DEVICE))

p_ntt = poly_ntt(p)

print(p)
print(p_ntt)

print("Shape:", p_ntt.coeffs.shape)
print("Device:", p_ntt.coeffs.device)
print("First 10:", p_ntt.tolist()[:10])
print("Min/max:", min(p_ntt.tolist()), max(p_ntt.tolist()))
```

```python
p = Poly(torch.arange(KYBER_N, dtype=torch.int16, device=DEVICE))

p_ntt = poly_ntt(p)
p_back = poly_invntt_tomont(p_ntt)

print("Original first 10:", p.tolist()[:10])
print("After NTT+InvNTT first 10:", p_back.tolist()[:10])
print("Min/max:", min(p_back.tolist()), max(p_back.tolist()))
```

```python
## Basemul testing
from mlkem_gpu.ntt_gpu import basemul


a = torch.tensor(
    [10, 20],
    dtype=torch.int16,
    device=DEVICE,
)

b = torch.tensor(
    [30, 40],
    dtype=torch.int16,
    device=DEVICE,
)

result = basemul(
    a,
    b,
    zeta=-1044,
)

print(result)
print(result.shape)
```
