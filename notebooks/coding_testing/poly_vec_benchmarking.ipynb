```python
import torch

from mlkem_gpu.device import DEVICE
from mlkem_gpu.params import KYBER_K, KYBER_N
from mlkem_gpu.poly_gpu import poly_ntt

from mlkem_gpu.polyvec_gpu import PolyVec, polyvec_basemul_acc_montgomery
```

```python
a = PolyVec(torch.arange(KYBER_K * KYBER_N, dtype=torch.int16, device=DEVICE))
b = PolyVec(torch.arange(KYBER_K * KYBER_N, dtype=torch.int16, device=DEVICE))

# Convert each polynomial in the polyvec to NTT domain
for i in range(KYBER_K):
    a.coeffs[i] = poly_ntt(type("TempPoly", (), {"coeffs": a.coeffs[i]})()).coeffs
    b.coeffs[i] = poly_ntt(type("TempPoly", (), {"coeffs": b.coeffs[i]})()).coeffs

r = polyvec_basemul_acc_montgomery(a, b)

print(r)
print("Shape:", r.coeffs.shape)
print("Device:", r.coeffs.device)
print("First 10:", r.tolist()[:10])
print("Min/max:", min(r.tolist()), max(r.tolist()))
```
