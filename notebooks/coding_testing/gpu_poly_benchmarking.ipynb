```python
import torch

from mlkem_gpu.device import DEVICE
from mlkem_gpu.params import KYBER_N

from mlkem_gpu.poly_gpu import Poly, poly_add, poly_sub, poly_reduce
from mlkem_gpu.polyvec_gpu import PolyVec, polyvec_add, polyvec_reduce

from mlkem_gpu.poly_gpu import (
    poly_ntt,
    poly_basemul_montgomery,
)
```

```python
a = Poly([3000] * 256)
b = Poly([1000] * 256)

c = poly_add(a, b)
d = poly_sub(a, b)

print(a)
print(b)

print("Add first 5:", c.tolist()[:5])
print("Sub first 5:", d.tolist()[:5])
```

```python
a = PolyVec([[3000] * 256, [1000] * 256])
b = PolyVec([[1000] * 256, [3000] * 256])

c = polyvec_add(a, b)

print(a)
print(b)
print(c)

print("Shape:", c.coeffs.shape)
print("First polynomial first 5:", c.tolist()[0][:5])
print("Second polynomial first 5:", c.tolist()[1][:5])
```

```python
a = Poly(torch.arange(KYBER_N, dtype=torch.int16, device=DEVICE))
b = Poly(torch.arange(KYBER_N, dtype=torch.int16, device=DEVICE))

a_ntt = poly_ntt(a)
b_ntt = poly_ntt(b)

r = poly_basemul_montgomery(a_ntt, b_ntt)

print(r)
print("Shape:", r.coeffs.shape)
print("Device:", r.coeffs.device)
print("First 10:", r.tolist()[:10])
print("Min/max:", min(r.tolist()), max(r.tolist()))
```
